### Counting the Number of Employees by Years

### Boxplot and Countplot of Years of Experience

```python
plt.figure(figsize=(20, 6))


plt.subplot(1,2,1)
plt.title("Boxplot of Years")
sns.boxplot(x="Years",data=df)



plt.subplot(1,2,2)
plt.title("Count of Years")
sns.countplot(df,x=df["Years"])

plt.tight_layout()
plt.show()
```

### Calculating the IQR and Identifying Outliers for Years

```python
years=df["Years"]
Q1 = years.quantile(0.25)
Q3 = years.quantile(0.75)


IQR = Q3 - Q1


lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR


print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")
```

### Filtering Data to Remove Outliers in Years

```python
not_outlier_year=df[(df["Years"]>lower_bound) & (df["Years"]<upper_bound)]
not_outlier_year
```

### Department vs. Years Distribution After Removing Outliers

```python
sns.countplot(not_outlier_year,x=not_outlier_year["Department"],hue=not_outlier_year["Years"])
plt.title("Department vs. Years Distribution")
plt.ylabel("Count")
plt.xlabel("Department")
plt.xticks(rotation=90)
plt.show()
```

### Department vs. Years Distribution (Including Outliers)

```python
sns.countplot(df,x=df["Department"],hue=df["Years"])
plt.title("Department vs. Years Distribution")
plt.ylabel("Count")
plt.xlabel("Department")
plt.xticks(rotation=90)
plt.show()
```

### Cross-tabulation of Department and Years After Removing Outliers

```python
Department_by_Years=pd.crosstab(not_outlier_year["Department"],not_outlier_year["Years"])
Department_by_Years
```
